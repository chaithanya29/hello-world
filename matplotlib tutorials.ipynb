```python
import matplotlib.pyplot as plt
import numpy as np


def f(t):
    'A damped exponential'
    s1 = np.cos(2 * np.pi * t)
    e1 = np.exp(-t)
    return s1 * e1


t1 = np.linspace(0.0, 5.0,100)

l1 = plt.plot(t1, f(t1), 'ro')
plt.setp(l1, markersize=3)
plt.setp(l1,markerfacecolor='C2')#setp sets up the various  parameters like the marker size , markercolour and all
plt.show(l1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2*np.pi, 0.02)
y = np.sin(x)
y1 = np.sin(2*x)
y2 = np.sin(3*x)
ym1 = np.ma.masked_where(y1 > 0.5, y1)
ym2 = np.ma.masked_where(y2 < -0.5, y2)

lines = plt.plot(x, y, x, ym1, x, ym2, 'o')
plt.setp(lines[0], linewidth=1)
plt.setp(lines[1], linewidth=1)
plt.setp(lines[2], markersize=1)

plt.legend(('No mask', 'Masked if > 0.5', 'Masked if < -0.5'),
           loc='upper right')
plt.title('Masked line demo')
plt.show()

```
